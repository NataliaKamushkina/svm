```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
from sklearn import datasets
%matplotlib inline
# 1. Получить данные 
# 2. сделать SVM Метод опорных векторовна этих данных ( использовать тот что в sklearn но только с параметром kernel='linear'); 
# посчитать ширину разделающей полосы и отрисовать ее

# 3. сделать kNN Метод k ближайших соседей и сравнить результат с SVM по метрикам

Метод опорных векторов - это метод машинного обучения, 
целью которого является попытка классифицировать входные наборы данных в один из двух классов.

Основным принципом метода ближайших соседей является то, что объект присваивается тому классу,
который является наиболее распространённым среди соседей данного элемента.
```

```python
n_samples = 4096

X, y = datasets.make_blobs(n_samples=n_samples, centers=3, n_features=2)


fig, ax = plt.subplots(1, 1, figsize=(15, 8))
colors = {0:'red', 1:'blue'}

for key, value in colors.items():   
    ax.scatter(X[y==key, 0], X[y==key, 1], c=value, label=key)

plt.axis('tight')

plt.show()
```

```python
from sklearn.svm import SVC
```

```python
n_samples = 1000

X, y = datasets.make_circles(n_samples=n_samples, noise=0.1, factor=0.3)

fig, ax = plt.subplots(1, 1, figsize=(15, 8))
colors = {0:'red', 1:'blue'}
for key, value in colors.items():   
    ax.scatter(X[y==key, 0], X[y==key, 1], c=value, label=key)
plt.axis('tight')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
model = SVC(kernel='linear') ### linear
model.fit(X, y)

#x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
#y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

predicted = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

random_indxs = np.random.randint(0, X.shape[0], 200)

plt.rcParams['figure.figsize'] = (10,8)
plt.pcolormesh(xx, yy, predicted, cmap='autumn')
plt.scatter(X[random_indxs, 0], X[random_indxs, 1], 
            c=y[random_indxs], s=50, cmap='autumn', edgecolors='black', linewidth=1.5)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
